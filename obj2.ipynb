```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('D:/diabetes+130-us+hospitals+for+years+1999-2008/cleaned_diabetic_data_with_Median.csv')

def convert_categorical_to_numeric(df, column_name):
    """Convert categorical column to numeric using label encoding."""
    unique_values = df[column_name].unique()
    value_map = {value: idx for idx, value in enumerate(unique_values)}
    df[column_name] = df[column_name].map(value_map)
    return df, value_map

df, metformin_map = convert_categorical_to_numeric(df, 'metformin')
df, glipizide_map = convert_categorical_to_numeric(df, 'glipizide')
df, glyburide_map = convert_categorical_to_numeric(df, 'glyburide')
df, insulin_map = convert_categorical_to_numeric(df, 'insulin')
df, readmitted_map = convert_categorical_to_numeric(df, 'readmitted')

print("\nMissing Values Summary:")
print(df[['metformin', 'glipizide', 'glyburide', 'insulin', 'readmitted']].isnull().sum())

numeric_columns = ['metformin', 'glipizide', 'glyburide', 'insulin', 'readmitted']
categorical_columns = df.columns.difference(numeric_columns)

df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

for column in categorical_columns:
    mode = df[column].mode()[0]  
    df[column].fillna(mode, inplace=True)


print("\nMissing Values After Handling:")
print(df[numeric_columns].isnull().sum())
print(df[categorical_columns].isnull().sum())

correlation_matrix = df[['metformin', 'glipizide', 'glyburide', 'insulin', 'readmitted']].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

print("correlation_matrix:")
print(correlation_matrix)

```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt
 
file_path = pd.read_csv('D:/diabetes+130-us+hospitals+for+years+1999-2008/cleaned_diabetic_data_with_Median.csv')
data = pd.read_csv(file_path)
 
label_encoder = LabelEncoder()
 
data['insulin_encoded'] = label_encoder.fit_transform(data['insulin'].astype(str))
data['max_glu_serum_encoded'] = label_encoder.fit_transform(data['max_glu_serum'].astype(str))
data['metformin_encoded'] = label_encoder.fit_transform(data['metformin'].astype(str))
data['readmitted_encoded'] = label_encoder.fit_transform(data['readmitted'].astype(str))
 
print("Columns in DataFrame:", data.columns)
 
numerical_features = [
    'num_lab_procedures', 'time_in_hospital', 'number_emergency',
    'num_medications', 'num_procedures', 'number_diagnoses',
    'insulin_encoded', 'max_glu_serum_encoded', 'metformin_encoded', 'readmitted_encoded'
]
 
missing_cols = [col for col in numerical_features if col not in data.columns]
if missing_cols:
    print("Missing columns:", missing_cols)
else:
    correlation_matrix = data[numerical_features].corr()
 
    plt.figure(figsize=(14, 10))
    sns.heatmap(correlation_matrix, annot=True, cmap='magma', fmt='.2f', vmin=-1, vmax=1, center=0, cbar_kws={'shrink': .75})
    plt.title('Correlation Matrix of Numerical Variables Including Encoded Variables')
    plt.show()
```
