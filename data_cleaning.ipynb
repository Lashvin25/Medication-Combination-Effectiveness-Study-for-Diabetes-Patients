```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('D:/diabetes+130-us+hospitals+for+years+1999-2008/bda_file2 1.csv')


columns_to_check = ['diag_1', 'diag_2', 'diag_3']


question_mark_counts = {}


for column_name in columns_to_check:
    question_mark_count = df[column_name].str.count(r'\?').sum()
    question_mark_counts[column_name] = question_mark_count

colors = ['#318CE7', '#051094', '#40E0D0']  

plt.bar(question_mark_counts.keys(), question_mark_counts.values(), color=colors)
plt.xlabel('Columns')
plt.ylabel("Count of '?'")
plt.title("diag_1, diag_2, and diag_3 Columns")
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('D:/diabetes+130-us+hospitals+for+years+1999-2008/bda_file2 1.csv')

columns_to_check = ['diag_1', 'diag_2', 'diag_3']

for column_name in columns_to_check:
    df[column_name] = pd.to_numeric(df[column_name], errors='coerce')

    mean_value = df[column_name].mean()
    median_value = df[column_name].median()
    
    print(f"Column '{column_name}':")
    print(f"  Mean: {mean_value}")
    print(f"  Median: {median_value}")
    print("\n" + "-"*50 + "\n")
```

```python
df.to_csv('D:/diabetes+130-us+hospitals+for+years+1999-2008/cleaned_diabetic_data_without_Median1.csv', index=False)
```

```python

median_values = {
    'diag_1': 440.0,
    'diag_2': 425.0,
    'diag_3': 403.0
}

for column_name in columns_to_check:
    df[column_name].replace('?', pd.NA, inplace=True)
    df[column_name] = pd.to_numeric(df[column_name], errors='coerce')
    df[column_name].fillna(median_values[column_name], inplace=True)
```

```python

df.to_csv('D:/diabetes+130-us+hospitals+for+years+1999-2008/cleaned_diabetic_data.csv', index=False)
```
