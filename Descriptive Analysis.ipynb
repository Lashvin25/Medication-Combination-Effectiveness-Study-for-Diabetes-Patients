Data Distribution

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('D:/diabetes+130-us+hospitals+for+years+1999-2008/cleaned_diabetic_data_with_Median.csv')

sns.set(style='whitegrid')

def plot_with_gradient(col, color1, color2):
    plt.figure(figsize=(10, 6))

    bars = sns.countplot(x=col, data=df, color=color1, edgecolor='black')

    for bar in bars.patches:
        bar.set_facecolor(color1)
        bar.set_edgecolor('black')
        bar.set_linewidth(1.5)

        bar.set_hatch('//')

        height = bar.get_height()
        bars.annotate(f'{int(height)}',
                      xy=(bar.get_x() + bar.get_width() / 2, height),
                      xytext=(0, 3), 
                      textcoords="offset points",
                      ha='center', va='bottom', fontsize=12, color='black')

    plt.title(f'Distribution of {col.capitalize()} Usage', fontsize=16)
    plt.xlabel(f'{col.capitalize()} Usage', fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

columns_and_colors = [
    ('metformin', 'skyblue', 'dodgerblue'),
    ('glipizide', 'salmon', 'firebrick'),
    ('glyburide', 'lightgreen', 'forestgreen'),
    ('insulin', 'orange', 'darkorange'),
    ('repaglinide', 'purple', 'indigo'),
    ('nateglinide', 'red', 'darkred'),
    ('chlorpropamide', 'cyan', 'darkcyan'),
    ('glimepiride', 'magenta', 'darkmagenta'),
    ('acetohexamide', 'lime', 'green'),
    ('tolbutamide', 'brown', 'saddlebrown'),
    ('readmitted', 'blue', 'navy'),
    ('diabetesMed', 'pink', 'deeppink')
]

for col, color1, color2 in columns_and_colors:
    plot_with_gradient(col, color1, color2)
```

Correlation Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv('D:/diabetes+130-us+hospitals+for+years+1999-2008/cleaned_diabetic_data_with_Median.csv')

def convert_categorical_to_numeric(df, column_name):
    """Convert categorical column to numeric using label encoding."""
    unique_values = df[column_name].unique()
    value_map = {value: idx for idx, value in enumerate(unique_values)}
    df[column_name] = df[column_name].map(value_map)
    return df, value_map

columns_to_convert = [
    'metformin', 'glipizide', 'glyburide', 'insulin', 
    'repaglinide', 'nateglinide', 'chlorpropamide', 
    'glimepiride', 'acetohexamide', 'tolbutamide', 
    'readmitted'
]

value_maps = {}
for column in columns_to_convert:
    df, value_map = convert_categorical_to_numeric(df, column)
    value_maps[column] = value_map

print("\n--- Missing Values Summary Before Handling ---")
print(df[columns_to_convert].isnull().sum())

df[columns_to_convert] = df[columns_to_convert].fillna(df[columns_to_convert].mean())
categorical_columns = df.columns.difference(columns_to_convert)
for column in categorical_columns:
    mode = df[column].mode()[0]
    df[column].fillna(mode, inplace=True)

print("\n--- Missing Values Summary After Handling ---")
print(df[columns_to_convert].isnull().sum())
print(df[categorical_columns].isnull().sum())

correlation_matrix = df[columns_to_convert].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, linecolor='black')
plt.title('Correlation Matrix of Selected Variables', fontsize=18, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.show()
print("\n--- Correlation Matrix ---")
print(correlation_matrix)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

df = pd.read_csv('D:/diabetes+130-us+hospitals+for+years+1999-2008/cleaned_diabetic_data_with_Median.csv')

readmitted_mapping = {
    '>30': 'Up',
    '<30': 'Down',
    'NO': 'No'
}
df['readmitted_mapped'] = df['readmitted'].map(readmitted_mapping)

medications = ['metformin', 'glipizide', 'glyburide', 'insulin', 'repaglinide', 'nateglinide',
                'chlorpropamide', 'glimepiride', 'acetohexamide', 'tolbutamide']

df['medication_combination'] = df[medications].fillna('No').apply(lambda row: ' & '.join(row), axis=1)

top_combinations = [
    ('metformin', 'nateglinide'),
    ('metformin', 'acetohexamide'),
    ('glipizide', 'nateglinide'),
    ('insulin', 'acetohexamide'),
    ('repaglinide', 'acetohexamide'),
    ('nateglinide', 'glimepiride')
]

def random_color_palette():
    colors = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]) for i in range(3)]
    return colors

for med1, med2 in top_combinations:
    df_filtered = df[df[med1].notna() & df[med2].notna()]
    df_filtered['medication_combination'] = df_filtered[med1].astype(str) + ' & ' + df_filtered[med2].astype(str)

    combinations = df_filtered['medication_combination'].unique()
    results = []

    for combo in combinations:
        subset = df_filtered[df_filtered['medication_combination'] == combo]
        percentages = subset['readmitted_mapped'].value_counts(normalize=True) * 100
        results.append({
            'Combination': combo,
            'Up': percentages.get('Up', 0),
            'Steady': percentages.get('Steady', 0),
            'No': percentages.get('No', 0)
        })

    results_df = pd.DataFrame(results).set_index('Combination')
    results_df_melted = results_df.reset_index().melt(id_vars='Combination', var_name='Status', value_name='Percentage')

    plt.figure(figsize=(14, 6))

    sns.barplot(data=results_df_melted, x='Combination', y='Percentage', hue='Status', palette=random_color_palette(), edgecolor='black')

    plt.xticks(rotation=90)
    plt.xlabel('Medication Combination', fontsize=12)
    plt.ylabel('Percentage (%)', fontsize=12)
    plt.title(f'Percentage of Readmitted Status for {med1} & {med2}', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)

    highest_up = results_df['Up'].max() if 'Up' in results_df else 0
    print(f'For combination of {med1} and {med2}:')
    print(f'Highest percentage of Up: {highest_up:.2f}%')
    print()

    plt.tight_layout()
    plt.show()
```
